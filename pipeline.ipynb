## - Get Artist info by top artist info

```python
# Get total number of top artists available
total_artists = sp.current_user_top_artists(limit=1, time_range='long_term')['total']

print(total_artists)
```

```python
all_top_artists = []
limit = 50

start = 0
stop = total_artists
step = limit

for offset in range(start, stop, step):
    top_artists = sp.current_user_top_artists(limit=limit, offset=offset, time_range='long_term')
    all_top_artists.extend(top_artists['items'])

print("=== Your Top Artists ===")
for idx, artist in enumerate(all_top_artists):
    print(f"{idx+1}. {artist['name']}")
```

```python
# print(json.dumps(all_top_artists, indent=2))
```

```python
top_artist_ranking = {artist['id']: rank + 1 for rank, artist in enumerate(all_top_artists)}
```

```python
# Top Artists are cached, call api if not found

from time import sleep

# Assumes this exists already:
artist_lookup = {artist['id']: artist for artist in all_top_artists}

artist_cache = {}

def get_artist_info(artist_id):
    # Check if cached
    if artist_id in artist_cache:
        return artist_cache[artist_id]

    # Check Top Artist List
    if artist_id in artist_lookup:
        artist_cache[artist_id] = artist_lookup[artist_id]
        return artist_cache[artist_id]

    # Fallback: call the API
    try:
        artist_info = sp.artist(artist_id)
        artist_cache[artist_id] = artist_info
        sleep(0.1)  # Avoid hitting rate limits
    except Exception as e:
        print(f"Error fetching artist {artist_id}: {e}")
        artist_cache[artist_id] = None

    return artist_cache[artist_id]
```

# Creating the CSV

```python
import csv

# Create the CSV header
csv_fields = [
    "Track Name", "Artist Name", "Album Name", "Album Release Date",
    "Track Duration", "Track Popularity", "Track Release Date", "Explicit",
    "ISRC", "Artist Genre", "Artist Popularity", "Artist Follower Count",
    "Album Picture", "Artist Picture", "Album ID", "Track ID", "Artist ID",
    "TopSongsRating", "TopArtistRanking"
]

rows = []

for idx, item in enumerate(all_top_tracks):
    track = item
    album = track['album']
    artist_id = track['artists'][0]['id']

    # Fetch artist data
    # artist_info = sp.artist(artist_id) # Rate Limit Problems
    artist_info = get_artist_info(artist_id)
    top_artist_rank = top_artist_ranking.get(artist_id)


    rows.append({
        "Track Name": track['name'],
        "Artist Name": track['artists'][0]['name'],
        "Album Name": album['name'],
        "Album Release Date": album['release_date'],
        "Track Duration": track['duration_ms'],
        "Track Popularity": track['popularity'],
        "Track Release Date": album['release_date'],  # Same as album
        "Explicit": track['explicit'],
        "ISRC": track['external_ids'].get('isrc'),
        "Artist Genre": ", ".join(artist_info.get('genres', [])),
        "Artist Popularity": artist_info.get('popularity'),
        "Artist Follower Count": artist_info['followers']['total'],
        "Album Picture": album['images'][0]['url'] if album.get('images') else None,
        "Artist Picture": artist_info['images'][0]['url'] if artist_info.get('images') else None,
        "Album ID": album['id'],
        "Track ID": track['id'],
        "Artist ID": artist_id,
        "TopSongsRating": idx + 1,  # Position in list
        "TopArtistRanking": top_artist_rank
    })
```

```python
with open("spotify_saved_tracks.csv", "w", newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=csv_fields)
    writer.writeheader()
    writer.writerows(rows)

print("CSV exported: spotify_saved_tracks_rebuilt.csv")
```

```python
# Fix Duration time to mm:ss

import pandas as pd

# Load CSV
df = pd.read_csv("spotify_saved_tracks.csv")

# Conversion function: ms → mm:ss
def ms_to_mmss(ms):
    seconds = int(ms // 1000)
    minutes = seconds // 60
    seconds = seconds % 60
    return f"{minutes}:{seconds:02d}"

# Overwrite the column
df['Track Duration'] = df['Track Duration'].apply(ms_to_mmss)

# Save and preview
df.to_csv("spotify_saved_tracks_updated.csv", index=False)
print(df[['Track Name', 'Track Duration']].head())
```

```python
# Patch in Date Added - Data
# Build mapping of Track ID → Date Added from saved tracks
added_at_lookup = {}
limit = 50
total_saved = sp.current_user_saved_tracks(limit=1)['total']

for offset in range(0, total_saved, limit):
    results = sp.current_user_saved_tracks(limit=limit, offset=offset)
    for item in results['items']:
        track_id = item['track']['id']
        added_at = item['added_at']
        added_at_lookup[track_id] = added_at
```

```python
import pandas as pd

# Load CSV
df = pd.read_csv("spotify_saved_tracks_updated.csv")

# Add new column using the lookup
df['Date Added'] = df['Track ID'].map(added_at_lookup)

# Save to new file
df.to_csv("spotify_saved_tracks_updated_with_date_added.csv", index=False)
```

```python
import pandas as pd
import sqlite3

# Load the CSV as a pandas dataframe
df = pd.read_csv("spotify_saved_tracks_updated_with_date_added_removed_album_date_lowercase_header_fixed.csv")

# Connect to / Create a SQLite database
conn = sqlite3.connect("spotify.db")  # or ":memory:" for in-RAM only

# Write the DataFrame to a new table - column names are automatically done - datatypes inferred
df.to_sql("spotify", conn, if_exists="replace", index=False)

# Querying DB
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print("Tables:", cursor.fetchall())

cursor.execute("SELECT * FROM spotify LIMIT 5;")
for row in cursor.fetchall():
    print(row)

# Close
conn.close()
```
